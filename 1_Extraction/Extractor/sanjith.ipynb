```python
#save all pages
import deepdoctection as dd
from IPython.core.display import HTML
from matplotlib import pyplot as plt
import os

analyzer = dd.get_dd_analyzer()  # instantiate the built-in analyzer similar to the Hugging Face space demo

df = analyzer.analyze(path="./sample/image.pdf")  # setting up pipeline
df.reset_state()  # Trigger some initialization

doc = iter(df)

# Create a directory to save the pages if it doesn't exist
output_dir = "./sample/pages"
os.makedirs(output_dir, exist_ok=True)

# Iterate through all the pages
for page_number, page in enumerate(doc, start=1):
    image = page.viz()

    # Display each page
    plt.figure(figsize=(25, 17))
    plt.axis('off')
    plt.imshow(image)
    plt.title(f"Page {page_number}")
    print(page.words)
    # Save the page as an image in the output directory
    image_filename = os.path.join(output_dir, f"page_{page_number}.png")
    plt.savefig(image_filename, bbox_inches='tight')
    plt.close()
    print(page.text)

# Notify the user about the saved pages
print(f"All pages saved in {output_dir}")
```

```python
import re
import json

data = {}
current_index = 0
mainTitleFlag = False
prev = None

for dp in df:
    layout_items = [layout for layout in dp.layouts if layout.reading_order is not None]
    layout_items.sort(key=lambda x: x.reading_order)

    for item in layout_items:
        if item.category_name == "title":
            if mainTitleFlag:
                data[current_index-1] = {"mainTitle": prev.text}
            data[current_index] = {"title": item.text}
            current_index += 1
            mainTitleFlag = True
            prev = item

        elif item.category_name in ["text", "list"]:
            if mainTitleFlag:
                mainTitleFlag = False
            data[current_index] = {"text": item.text}
            current_index += 1

with open("output.json", 'w') as json_file:
    json.dump(data, json_file, indent=4)
```
