```python
import json
import copy
from google.colab import files

# Function to summarize text content
def generate_summary(model, text_content, ratio=0.5):
    if ratio < 0 or ratio > 1:
        ratio = 0.5

    prompt = f"Summarize the following text to reduce it to {ratio} of the original text:\n" + text_content
    response = model.generate_content(
        contents=prompt
    )
    return response.text

# Create the model
model = genai.GenerativeModel('gemini-pro')

# Function to process and summarize the JSON file
def process_json_file(file_path, output_file_name, ratio=0.5):
    with open(file_path, 'r') as file:
        data = json.load(file)

    # Create a new JSON structure for the updated data
    updated_json = {}

    summary=""
    # Iterate over the JSON data
    for key, obj in data.items():
        # Check if the current object is a 'text' object for summarization
        if 'text' in obj:
            # Generate the summary for the 'text' object
            summary = generate_summary(model, obj['text'], ratio)
            # Update the 'text' object in the new JSON structure with the summary
            updated_json[key] = {'text': summary}
        else:
            # Directly copy other objects without changes
            updated_json[key] = obj


    # Save the updated content to a new JSON file
    with open(output_file_name, 'w') as file:
        json.dump(updated_json, file, indent=4)

# Upload the JSON file
uploaded = files.upload()
file_name = next(iter(uploaded))

# Input the ratio
ratio = float(input("Enter the ratio to reduce the text to (between 0 and 1): "))

# Process the JSON file and generate the output
output_file_name = 'summarized_output.json'
process_json_file(file_name, output_file_name, ratio)

# Download the summarized output JSON file
files.download(output_file_name)
```
